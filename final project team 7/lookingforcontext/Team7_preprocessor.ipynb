```python
import os
from transformers import pipeline

context = open(f"pg2097.txt", 'r',encoding="utf8")
qa_model = pipeline("question-answering")
test = []
question = "What was Jonathan Small's role?" 
#initializing
f = open("What_was_Jonathan_Smalls_role.txt", "a",buffering=1)

index = 0
lineNumber = 0
testContext = context.read()
textContext = testContext.split()
for word in textContext:
    #print(str(index) + " " + word)
    if(index < 300):
        test.append(word)
        s = ' '.join(test)
    else:
        test.append(word)
        test.pop(0)
        s = ' '.join(test)
        response = qa_model(question = question, context = s)
        f.write(f"{response['answer']} ({response['score']} Line number: {lineNumber} " + test[0-20])
        f.write("\n")
        
    lineNumber +=1
    index += 1

f.close()
print("done")


```
